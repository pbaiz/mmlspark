# Setup

```python
import pandas as pd
import numpy as np
```

# Import Data

```python
df_benchmark = pd.read_csv('datasets/bank-additional-full-kaggle-dataset-final.csv', sep = ',')
df_y = pd.read_csv('datasets/bank-additional-full-kaggle-dataset-y.csv', sep = ',')
df_benchmark.head()
```

```python
df_y.head()
```

```python
df_benchmark = pd.concat([df_benchmark, df_y], axis=1)
```

```python
df_benchmark.head()
```

```python
df_benchmark = df_benchmark.rename({"['y']_yes": 'y'}, axis=1)
```

```python
df_benchmark.head()
```

```python
df_benchmark = df_benchmark.drop('duration', axis=1)
```

```python
df_benchmark.head()
```

```python
neg, pos = np.bincount(df_benchmark['y'])
total = neg + pos
print('Examples:\n    Total: {}\n    Positive: {} ({:.2f}% of total)\n'.format(
    total, pos, 100 * pos / total))
```

## Data processing

```python
# drop rows with missing values
df_benchmark = df_benchmark.dropna()
```

```python
df_benchmark.shape
```

```python
# extract labels
df_y = df_benchmark[["y"]]
df_final = df_benchmark.drop("y", 'columns')
```

```python
df_final.head()
```

```python
df_y.head()
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(df_final, df_y, test_size = 0.1942313295, random_state = 101)
```

```python
print("############# Data summary #############")
print(f"x_train has shape: {x_train.shape}")
print(f"y_train has shape: {y_train.shape}")
print(f"x_test has shape: {x_test.shape}")
print(f"y_test has shape: {y_test.shape}")
print("#######################################")
```

```python
x_train.to_csv('datasets/bank-additional-full-x-train.csv', index=False)
y_train.to_csv('datasets/bank-additional-full-y-train.csv', index=False)
x_test.to_csv('datasets/bank-additional-full-x-test.csv', index=False)
y_test.to_csv('datasets/bank-additional-full-y-test.csv', index=False)
```
